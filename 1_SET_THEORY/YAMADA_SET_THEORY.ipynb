```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt
%matplotlib inline

def drawVenn2(A,B,labels = ('A', 'B')):
    '''
    listing elements in Venn diagram of two sets
    '''
    v2=venn2([A,B],set_labels = labels);
    v2.get_label_by_id('01').set_text(', '.join(map(str,B-A)))
    v2.get_label_by_id('10').set_text(', '.join(map(str,A-B)))
    if A&B:
        v2.get_label_by_id('11').set_text(', '.join(map(str,A&B)))

def drawVenn3(A,B,C,labels = ('A', 'B', 'C')):
    '''
    listing elements in Venn diagram of three sets, A, B, C
    '''
    v3=venn3([A,B,C],set_labels = labels);
    if A-B-C:
        v3.get_label_by_id('100').set_text('\n'.join(map(str,A-B-C)))
        v3.get_patch_by_id('100').set_alpha(0.6)
        v3.get_patch_by_id('100').set_facecolor('Yellow')
    if A&B-C:
        v3.get_label_by_id('110').set_text('\n'.join(map(str,A&B-C)))
        v3.get_patch_by_id('110').set_alpha(0.3)
        v3.get_patch_by_id('110').set_facecolor('Red')
    if B-C-A:
        v3.get_label_by_id('010').set_text('\n'.join(map(str,B-C-A)))
        v3.get_patch_by_id('010').set_alpha(0.6)
        v3.get_patch_by_id('010').set_facecolor('Magenta')
    if A&C-B:
        v3.get_label_by_id('101').set_text('\n'.join(map(str,A&C-B)))
        v3.get_patch_by_id('101').set_alpha(0.3)
        v3.get_patch_by_id('101').set_facecolor('Green')
    if A&B&C:
        v3.get_label_by_id('111').set_text('\n'.join(map(str,A&B&C)))
        v3.get_patch_by_id('111').set_alpha(0.1)
        v3.get_patch_by_id('111').set_facecolor('Black')
    if B&C-A:
        v3.get_label_by_id('011').set_text('\n'.join(map(str,B&C-A)))
        v3.get_patch_by_id('011').set_alpha(0.3)
        v3.get_patch_by_id('011').set_facecolor('Blue')
    if C-B-A:
        v3.get_label_by_id('001').set_text('\n'.join(map(str,C-B-A)))
        v3.get_patch_by_id('001').set_alpha(0.6)
        v3.get_patch_by_id('001').set_facecolor('Cyan')
```

Show that:

1. $ A \setminus B \ = A \cap B^\complement $ \
if $ X \in A$ then $X \notin B $ \
if $ X \notin B$ then $X \in B^\complement $ \
if $ X \in B^\complement$ then $X \in A \land X \in B^\complement $\
if $ X \in A \land X \in B^\complement$ then $X \in A \cap B^\complement $ \
Then that means $ A \setminus B = A \cap B^\complement $ \

2. $ B = (B \cap A) \cup (B \cap A^c) $ \
Here, $ B = B $ \
if $ X \in (B \cap A)$ then $ X \in B \land X \in A $ \
if $ X \in B \land X \in A $, then $ X \in B $ \
if $ X \in (B \cap A^c) $, then $ X \in B \land X \in A^c $ \
if $ X \in B \land A^c $, then $ X \notin A $ \
In both cases, $ X \in B $, so $(B \cap A) \cup (B \cap A^c) = B $

3. $ B\setminus A = B \cap A^c $\
if $ X \in B $ and $ B\setminus A $, then $ X \notin A $\
if $ X \notin A $, then $ X \in A^c $\
if $ X \in B \land X \in A^c $, then $ B\setminus A = B \cap A^c $

4. $ A \cup B = A \cup (B \cap A^c) $\
if $ X \in A \cup B $, then it is true that either $ X \in A $ or $ X \in B $\
if $ X \in A $ or $ X \in B $, then certainly, $ X \in A $\
if $ X \in (A \cup (B \cap A^c)) $, then it is true that $ X \in A $ or $ X \in B $\
if $ X \in A $ or $ X \in B $, then certainly, $ X \in A $\
if $ X \notin A^c $, then $ (B \cap A^c) = B $ \

Now, let's prove the given statement:

**Proof**:

Let's start with the left-hand side: $(\left(\bigcup_{i=1}^{\infty}A_i^c\right)^c)$. This represents the complement of the union of the complements of sets $(A_i)$.

1. **De Morgan's Law for Complements** states that the complement of a union is equal to the intersection of the complements:

   $(\left(\bigcup_{i=1}^{\infty}A_i^c\right)^c = \bigcap_{i=1}^{\infty}(A_i^c)^c)$

2. Now, recall that the complement of a complement is the set itself:

   $(\left(\bigcup_{i=1}^{\infty}A_i^c\right)^c = \bigcap_{i=1}^{\infty}A_i)$
 
So, we have successfully shown that $(\left(\bigcup_{i=1}^{\infty}A_i^c\right)^c = \bigcap_{i=1}^{\infty}A_i)$, based on De Morgan's law for complements.

In a Survey of 120 People, it was found that:
- 65 read Newsweek Magazine
- 45 read Times Magazine
- 42 read Fortune Magazine
- 20 read Both Newsweek and Time
- 25 read both Newsweek and Fortune
- 15 read both Time and Fortune
- 8 read all three magazines
With these information do the following:<br>
1. Draw the venn diagram using python.
2. Find the number of people who read at $\textbf{"least one"}$ of the three magazines.
3. Find the number of people who read exactly one magazine.
<br>
<br>

```python
n_only = 28 
t_only = 18 
f_only = 10 
n_and_t = 12
n_and_f = 17
t_and_f = 7
all_three = 8

X = 65 + 45 + 42 - 20 - 25 - 15 + 8

f_readers = 42 - 25 - 15 + 8
n_readers = 65 - 25 - 20 + 8
t_readers = 45 - 15 -20 + 8
Y = f_readers + n_readers + t_readers

venn = venn3 (subsets=(n_only, t_only, n_and_t, f_only, n_and_f, t_and_f, all_three), set_labels=('NM', 'TM', 'FM'))

venn_circles = venn3_circles(subsets=(n_only, t_only, n_and_t, f_only, n_and_f, t_and_f, all_three))

venn_circles[0].set_lw(1.0)  
venn_circles[1].set_lw(1.0) 
venn_circles[2].set_lw(1.0)  

print("Number of people who read at least one of the 3 magazines:", X)
print("Total number of people who read exactly one magazine:", Y)

plt.show()
```

-  Where NM is A, TM is B, and FM is C.

    = 65 + 45 + 42 - 20 - 25 - 15 + 8 = 100 <br>
    = $A \cup B \cup C = 100$\
    = This means that:   **100 out of 120 people read at least one of the three magazines**
  
- $A \cap B \cap \ C = 8$ <br> (People who read all three magazines)
  
- $A \cup B - A \cap B \cap C = 20 - 8 = 12$ <br> (People who read only Newsweek and Times)
  
- $A \cup C - A \cap B \cap C = 25 - 8 = 17$ <br> (People who read only Newsweek and Fortune)
  
- $B \cup C - A \cap B \cap C = 15 - 8 = 7$ <br> (People who read only Times and Fortune)
  
- $A - A \cap B - A \cap C + A \cap B \cap C = 65 - 20 - 25 + 8 = 28$ <br> (People who read only Newsweek)
  
- $B - A \cap B - B \cap C + A \cap B \cap C = 45 - 20 - 15 + 8 = 18$ <br>  (People who read only Times)
